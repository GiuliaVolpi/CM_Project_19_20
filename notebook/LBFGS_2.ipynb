```python
%load_ext autoreload
%reload_ext autoreload
%autoreload 2
import sys
from os import path
sys.path.insert(0, "../ISANet/")
sys.path.insert(0, "./")

from isanet.model import Mlp
from isanet.optimizer import SGD, NCG, LBFGS
from isanet.datasets.monk import load_monk
from isanet.utils.model_utils import printMSE, printAcc, plotHistory
import matplotlib.pyplot as plt
import numpy as np
```

```python
def get_fitted_model(X_train, Y_train, optimizer, n_seed = 189, verbose = 1):
    np.random.seed(seed=n_seed)
    print("Build the model")
    model = Mlp()
    model.add(4, input= 17, kernel_initializer = 0.003, kernel_regularizer = 0.001)
    model.add(1, kernel_initializer = 0.003, kernel_regularizer = 0.001)

    model.set_optimizer(optimizer)

    model.fit(X_train,
            Y_train, 
            epochs=1000, 
            verbose=verbose)

    return model
```

# MONK 3

```python
print("Load Monk DataSet")
X_train, Y_train = load_monk("3", "train")

seed = 206 #623 #189
ln_maxiter = 10
results = []
```

```python
#############################
#          LBFGS m=3
#############################

optimizer = LBFGS(m=3, c1= 1e-4, c2=0.8, ln_maxiter = 100, l_eps=1e-6)

model = get_fitted_model(X_train, Y_train, optimizer, seed, 2)
h_lbfgs_m_3 = model.history 
```

```python
#############################
#          LBFGS m=5
#############################

optimizer = LBFGS(m=5, c1= 1e-4, c2=0.9, ln_maxiter = 100, l_eps=1e-6)

model = get_fitted_model(X_train, Y_train, optimizer, seed, 1)
h_lbfgs_m_5 = model.history 
```

```python
#############################
#          LBFGS m=10
#############################

optimizer = LBFGS(m=10, c1= 1e-4, c2=0.8, ln_maxiter = 100, l_eps=1e-6)

model = get_fitted_model(X_train, Y_train, optimizer, seed, 1)
h_lbfgs_m_10 = model.history 
```

```python
plt.plot(h_lbfgs_m_3["loss_mse"], linestyle='-')
plt.plot(h_lbfgs_m_5["loss_mse"], linestyle = '--')
plt.plot(h_lbfgs_m_10["loss_mse"], linestyle='-.')
plt.title('Monk1 - seed={}'.format(seed))
plt.ylabel("Loss")
plt.xlabel('Iteration')
plt.grid()
plt.yscale('log')
plt.legend(['LBFGS m=3','LBFGS m=5'.format(15),'LBFGS m=10'.format(3)], loc='upper right', fontsize='large')
plt.savefig('./plot/monk1_fr_56.eps', format='eps')
plt.show()
```

```python
plt.plot(h_lbfgs_m_3["loss_mse"])
plt.title('Monk3 - seed={}'.format(seed))
plt.ylabel("Loss")
plt.xlabel('Iteration')
plt.grid()
plt.yscale('log')
plt.legend(['LBFGS m=3','LBFGS m=5'.format(15),'LBFGS m=10'.format(3)], loc='lower left', fontsize='large')
plt.savefig('./plot/monk1_fr_56.eps', format='eps')
plt.show()
```
